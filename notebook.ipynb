```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data/imdb_movie_reviews.csv')
df.sample(10)
```

```python
df = df.sample(5000) # Reduced the sample size to accelerate training
```

```python
df.shape
```

# Text Cleaning Steps

* Remove html tags
* Remove special characters
* Convert every row to lowercase
* Remove stop words
* Stemming
* Vectorization

```python
# Step 01
def remove_html(text):
    return re.sub(re.compile('<.*?>'), '', text)
```

```python
df['review'] = df['review'].apply(remove_html)
```

```python
df.sample(10)
```

```python
# Step 02
def remove_special(text):
    x = ''
    for c in text:
        if c.isalnum():
            x += c
        else:
            x += ' '
    return x
```

```python
df['review'] = df['review'].apply(remove_special)
```

```python
# Step 03
def convert_lower(text):
    return text.lower()
```

```python
df['review'] = df['review'].apply(convert_lower)
```

```python
# Step 04
def remove_stopwords(text):
    x = []
    for c in text.split():
        if c not in stopwords.words('english'):
            x.append(c)
    y = x[:]
    x.clear()
    return y
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df.sample(10)
```

```python
# Step 05
ps = PorterStemmer()
y = []
def stem_words(text):
    for c in text:
        y.append(ps.stem(c))
    z = y[:]
    y.clear()
    return z
```

```python
df['review'] = df['review'].apply(stem_words)
```

```python
def join_back(lst):
    return ' '.join(lst)
```

```python
df['review'] = df['review'].apply(join_back)
```

```python
df.sample(10)
```

```python
# Step 06
cv = CountVectorizer(max_features=1000) # We tune the max_features parameter for increased accuracy
X = cv.fit_transform(df['review']).toarray()
```

```python
X.shape
```

```python
y = df.iloc[:, -1].values
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
y_pred3 = clf3.predict(X_test)
```

```python
print('GaussianNB:', accuracy_score(y_test, y_pred1))
print('MultinomialNB:', accuracy_score(y_test, y_pred2))
print('BernoulliNB:', accuracy_score(y_test, y_pred3))
```

```python
# Increasing accuracy through experimentation
feature_values = range(1000, 10000, 1000)
scores = {}

for feature in feature_values:
    cv = CountVectorizer(max_features=feature)
    X = cv.fit_transform(df['review']).toarray()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
    
    clf = MultinomialNB().fit(X_train, y_train) # You can choose the NB model of your choice
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    scores[feature] = accuracy
```

```python
for feature, accuracy in scores.items():
    print(f'Max Features: {feature}, Accuracy: {accuracy}')
```

```python
# Visualizing the above results
plt.plot(feature_values, list(scores.values()), marker='o', linestyle='--', color='b', label='Accuracy')
plt.title('Performance of NB model for Different Feature Values')
plt.xlabel('Number of Features')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
